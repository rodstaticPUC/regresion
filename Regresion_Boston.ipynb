# Data BOSTON: Regresion


```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import statistics
import warnings
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message')
```

```python
boston = load_boston()
X = boston.data
y = boston.target
x = range(0, boston.data.shape[0],1)
CRIME = 0; # tasa de delincuencia per capita por ciudad
ZN = 1; # proporcion de terreno residencial dividido en zonas para lotes de mas 25.000 pies cuadrados
INDUSTRIAL = 2; # proporcion de acres comerciales no minoristas por ciudad
ROOMS= 5; # numero medio de habitaciones por vivienda
DISTANCE=7 # distancias medias a centros de empleo de Boston
# boston.data.shape[0] # numeros de observaciones
#boston.data.shape[1] # numero de variables
print(X.shape) # dimension de la data
```

# Graficos de caracteristicas

```python
plt.figure(figsize=(30,8))
plt.plot(x,X[:, INDUSTRIAL], 'r-', label='Industrial')
plt.plot(x,X[:, DISTANCE], 'b-', label='Distancia')
plt.plot(x,X[:, ROOMS], 'g-', label='Habitaciones')
plt.rcParams.update({'font.size':18});
plt.legend(prop={'size':22})
```

**Calculo de la media y la varianza de cada caracteristica**

```python
mean = []; variance =[]

for i in range (boston.data.shape[1]):
    mean.append(statistics.mean(X[:,i]))
    variance.append(statistics.variance(X[:,i]))
```

Obtener cada media y varianza de cada caracteristica

```python
for i in range(boston.data.shape[1]):
    print(format(i) + "mean:{:.2f}".format(mean[i]) + ",variance:{:.2f}".format(variance[i]))
```

```python
fig,  axs = plt.subplots(1,2, figsize=(8,4))

axs[1]= Axes3D(fig,elev=30, azim=-40)
axs[1].scatter(X[:,DISTANCE],X[:,ROOMS],y,c='r',marker="o")
axs[1].set_xlabel("Distancia", fontsize=12)
axs[1].set_ylabel("Habitaciones", fontsize=14)
axs[1].set_zlabel("Boston", fontsize=14)
plt.show()
```
